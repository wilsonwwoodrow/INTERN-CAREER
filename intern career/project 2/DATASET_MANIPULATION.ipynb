```python
import matplotlib.pyplot as plt

# Sample data
unique_set = {
    'status': 1,
    'bed': 28,
    'bath': 23,
    'acre_lot': 1582,
    'city': 2623,
    'state': 12,
    'zip_code': 3264,
    'house_size': 5364,
    'prev_sold_date': 9479,
    'price': 3690
}

# Convert the unique_set to a Pandas Series
unique_counts = pd.Series(unique_set)

# Plot a bar chart to visualize the counts
plt.figure(figsize=(10, 6))
unique_counts.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Unique Counts for Each Attribute')
#Replace Count and Attribute with your required titles.
plt.xlabel('Count')

plt.ylabel('Attribute')
plt.show()
```

## Show all unique Data Sets for a particulat column

```python
# Specify the column for which you want to print unique values
column_name = 'city'

# Get unique values for the specified column
unique_values = data[column_name].unique()

# Print the unique values
print(f"Unique values for {column_name}:")
print(unique_values)
```

```python
len(unique_values)
```

```python
value_counts = data[column_name].value_counts()

# Plot a bar chart
plt.bar(value_counts.index, value_counts.values, color='skyblue')
plt.title(f'Occurrence of Unique Values in {column_name}')
plt.xlabel(column_name)
plt.ylabel('Count')
plt.show()
```
